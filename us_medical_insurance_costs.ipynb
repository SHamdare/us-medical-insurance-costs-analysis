# U.S. Medical Insurance Costs

**Import Libraries**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load the Dataset**

```python
df = pd.read_csv("/content/sample_data/insurance.csv")

df.head()
```

**Check it loaded correctly**

```python
df.info()
```

```python
df.describe()
```

**Check Missing Values**

```python
df.isnull().sum()
```

**Average Insurance Cost**

```python
average_cost = df["charges"].mean()

print("Average insurance cost:", average_cost)
```

**Smoking vs Insurance Charges**

```python
sns.boxplot(x="smoker", y="charges", data=df)

plt.title("Insurance Charges for Smokers vs Non-Smokers")

plt.show()
```

**Age vs Insurance Charges**

```python
sns.scatterplot(x="age", y="charges", data=df)

plt.title("Age vs Insurance Charges")

plt.show()
```

**BMI vs Insurance Charges**

```python
sns.scatterplot(x="bmi", y="charges", data=df)

plt.title("BMI vs Insurance Charges")

plt.show()
```

**Average Charges by Region**

```python
df.groupby("region")["charges"].mean()
```

**Average Charges by Gender**

```python
df.groupby("sex")["charges"].mean()
```

**Smoker Impact**

```python
df.groupby("smoker")["charges"].mean()
```

**Correlation Matrix**

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.title("Correlation Matrix")

plt.show()
```

**Conclusion **

Key Findings:

1. Smokers have significantly higher insurance costs.
2. Insurance charges increase with age.
3. Higher BMI tends to increase insurance costs.
4. Regional differences are relatively small.
